```python
import music21 as m21

# load file as a music21 stream
s = m21.converter.parse('BachJS-BWV846-1.mxl')
s.show('t')
```

```python
# we just need the pitch of each note
f = s.flat.notes
# f is a list of notes
f.show('t')
```

```python
# let's examine a note
n = f[1]
n.show('t')
print('n.pitches: ', n.pitches)
print('n.pitch: ', n.pitch) # note E in octave 4
print('n.pitch.midi: ', n.pitch.midi) # midi pitch
print('pitch class: ', n.pitch.midi%12) # pitch class
```

```python
# let's get all pitch classes:
pcs = []
for p in f.pitches:
  pcs.append( p.midi%12 )
print('pcs:', pcs)
```

```python
import numpy as np
# let's make the pitch class profile
pcp = np.zeros(12)
for p in pcs:
  pcp[ p ] += 1

print('unnormalised pcp: ', pcp)
# we don't care about how many notes the piece has
# we care about percentages -> %
# normalized
pcp = pcp / np.sum(pcp)
print('normalised pcp: ', pcp)
```

```python
# estimate tonality
p = m21.analysis.discrete.KrumhanslSchmuckler()
tonality = p.getSolution(s)
print('estimated tonality: ',  tonality)
```
